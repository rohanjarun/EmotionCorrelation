```python
#import dependencies/packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load/read in the CSV file
file_path = r'C:\Users\rohan.arun\Downloads\random_names_data.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head(10))
```

```python
# Display basic information about the dataset
print("Basic Information:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())

# Plotting histograms for each emotion score
emotions = ["Happiness", "Sadness", "Anger", "Fear", "Anxiousness", "Disgust"]
```

```python
# Boxplot for age distribution by gender
plt.figure()
df.boxplot(column='Age', by='Gender')
plt.title('Boxplot of Age by Gender')
plt.suptitle('')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.show()
```

```python
# Calculate and print the mean scores for each emotion by gender
mean_scores_by_gender = df.groupby('Gender')[emotions].mean()
print("\nMean Emotion Scores by Gender:")
print(mean_scores_by_gender)
```

```python
# Calculate the correlation matrix
emotions = ["Happiness", "Sadness", "Anger", "Fear", "Anxiousness", "Disgust"]
correlation_matrix = df[emotions].corr()

print("\nCorrelation Matrix:")
print(correlation_matrix)
```

```python
# Plot the correlation matrix as a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Emotions')
plt.show()
```
