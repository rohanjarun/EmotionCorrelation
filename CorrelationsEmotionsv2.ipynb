```python
#import dependencies/packages
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
# Load/read in the CSV file
file_path = r'C:\Users\rohan.arun\Downloads\single_person_emotions_varied_trends.csv'
df = pd.read_csv(file_path)
```

```python
print(df.head(10))
```

```python
# Display basic information about the dataset
print("\nBasic Information:")
print(df.info())

print("\nSummary Statistics:")
print(df.describe())
```

```python
# Plotting histograms for each emotion score
emotions = ["Calm", "Happy", "Pride", "Love", "Surprise", "Fear", "Anger", "Pensiveness", "Disgust"]
```

```python
for emotion in emotions:
    plt.figure()
    plt.hist(df[emotion], bins=20, edgecolor='black')
    plt.title(f'Histogram of {emotion} Scores')
    plt.xlabel('Score')
    plt.ylabel('Frequency')
    plt.show()
```

```python
# Calculate and plot the correlation matrix
correlation_matrix = df[emotions].corr()

print("\nCorrelation Matrix:")
print(correlation_matrix)

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix of Emotions')
plt.show()
```

```python
# Scatter plots to visualize relationships between emotions
sns.pairplot(df[emotions])
plt.suptitle('Pair Plot of Emotions', y=1.02)
plt.show()
```

```python
# Create a time index representing 60 seconds
df['Time'] = range(1, 61)
```

```python
# Plot the line graph
plt.figure(figsize=(14, 8))
```

```python
for emotion in emotions:
    plt.plot(df['Time'], df[emotion], label=emotion)

# Adding titles and labels
plt.title('Emotion Levels Over Time (60 seconds)')
plt.xlabel('Time (seconds)')
plt.ylabel('Emotion Score')
plt.legend(title='Emotions')
plt.grid(True)
plt.show()
```

```python
# Plot the line graph with moving averages
fig, ax = plt.subplots(figsize=(20, 8))  # Increased width to 20 inches

for emotion in emotions:
    ax.plot(df['Time'], df[emotion], label=emotion, alpha=0.5)  # Original data with lower opacity

# Adding titles and labels
ax.set_title('Emotion Levels Over Time (60 seconds)')
ax.set_xlabel('Time (seconds)')
ax.set_ylabel('Emotion Score')
ax.legend(title='Emotions', bbox_to_anchor=(1.05, 1), loc='upper left')
ax.grid(True)
plt.show()
```
